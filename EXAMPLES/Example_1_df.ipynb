```python
from importlib_resources import files

from pyLHCO.pylhco import read_lhco
```

```python
ref = files("pyLHCO")
DATA_PATH = str(ref)
```

```python
particles = read_lhco(DATA_PATH + "/dataset/example.lhco", OutputType = 'df')
```

```python
particles
```

Now you can play with the pandas data frame!

For example let's try to keep all the photons with $p_{t} > 10$ GeV and sum the $p_{t}$

```python
conds = ["particles['typ'] == 0", # type = 0 are photons
         "(particles['pt'] > 10)"]

evs = particles.copy()
for cond in conds:
    evs = evs[eval(cond)]

evs.groupby('evNum').sum()['pt']
```

Now let's do something a bit more complicated...
We want to check which events has at least 2 electrons, and from those count how many photons they have.

```python

# prompt: I want to group by evNum, then select the groups with at least 3 rows with typ = 1, and from them count how many rows has typ = 0.

import pandas as pd

# Group by 'evNum'
grouped = particles.groupby('evNum')

# Filter groups with at least 3 rows where 'typ' == 1
filtered_groups = grouped.filter(lambda x: (x['typ'] == 1).sum() >= 2)

# Count rows where 'typ' == 0 in the filtered groups
result = filtered_groups[filtered_groups['typ'] == 0].groupby('evNum').size()

result
```

```python
# prompt: group by evNum, select those which has at least 2 rows with typ = 1, compute the sum of pt for those rows, select the groups which has a sum bigger than 10, and then count how many rows with typ = 3 they have

# Group by 'evNum'
grouped = particles.groupby('evNum')

# Filter groups with at least 2 rows where 'typ' == 1 and sum pt > 10
filtered_groups = grouped.filter(lambda x: (x['typ'] == 1).sum() >= 2)
sum_pt_groups = filtered_groups[filtered_groups["typ"] == 1].groupby("evNum").sum()
filtered_sum_pt = sum_pt_groups[sum_pt_groups["pt"] > 10]

#get the evNums that meet the criteria
evNums = filtered_sum_pt.index

#select the rows from the original dataframe that match the evNums
final_df = filtered_groups[filtered_groups["evNum"].isin(evNums)]

#count how many rows with typ == 3 each group has
result = final_df[final_df['typ'] == 3].groupby('evNum').size()

result
```
